```python
import pandas as pd
import re

# Load the Excel file
df = pd.read_excel("Job opportunities.xlsx")

# Define skill keywords to search for
skill_keywords = [
    'SQL', 'Python', 'Excel', 'Power BI', 'SSRS', 'Flask', 'Django',
    'machine learning', 'REST API', 'data visualization', 'AWS',
    'Azure', 'Java', 'Cisco','Business Analysis','Consulting',
    'Architecture','QA','Windows','Agile', 'Project management',
]

# Function to extract skills from job description text
def extract_skills(text):
    if pd.isna(text):
        return ""
    found = [skill for skill in skill_keywords if re.search(skill, text, re.IGNORECASE)]
    return ", ".join(found)

# Apply the function and print results without adding a new column
for idx, row in df.iterrows():
    job_title = row.get("Job Title", "")
    job_description = row.get("Job Description", "")
    extracted = extract_skills(job_description)
    print(f"Job Title: {job_title}")
    print(f"Job Description: {job_description}")
    print(f"Extracted Skills: {extracted}")
    print("-" * 60)
```

```python
df.to_csv("job_skills_cleaned.csv", index=False)
```
